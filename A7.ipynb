```python
#Importamos las librerias:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#checamos si esta cargada la file
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(3)
```

```python
#Q1
sns.histplot(x = df['diastol'],
             bins=15,
           ).set_title('Histplot de diastol')

plt.show()
```

```python
#Q2
sns.lmplot(x='sistol', 
           y='peso', 
           data = df,
           hue='sexo',
           fit_reg = False, 
          )
```

```python
#Q3
sns.boxplot(x = df['sexo'],
            y = df['cintura'],
           ).set_title('Box plot de sexo - cintura')
plt.show()
```

```python
#Q4
sns.lmplot(x='peso', 
           y='cintura', 
           data = df,
           fit_reg = False,
          )
```

```python
#Q5 intento
sns.histplot(x = df['cintura'],
             y= df ['sexo'],
             bins=15,
           ).set_title('Histplot de diastol')

plt.show()
```

```python
#No sirve
sns.histplot(data=df.loc[df["sexo"]=="M",
                       ["cintura"]])
                        
plt.xticks(rotation=45, fontsize=17)
plt.yticks(fontsize=17)
plt.title("Box plot representation of Male data", fontsize=15)
plt.show()
```

```python
#No sirve
fig= plt.figure()
ax1= fig.add_subplot(3,2,1) #of rows,#of columns position

sns.histplot(data = df.loc[df["sexo"]=="1",
                         ["cintura"]
                         ]
           )
ax2= fig.add_subplot(3,2,2) #of rows,#of columns position

sns.histplot(data = df.loc[df["sexo"]=="2",
                         ["cintura"]
                         ]
           )
ax1.title.set_text("Male box plot")
ax2.title.set_text("Female box plot")

plt.show()


```
