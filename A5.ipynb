```python
#Importamos las librerias:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#checamos si esta cargada la file
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(3)
```

```python
#Q1 box plot de edad
sns.boxplot(x = df['edad'],     
           ).set_title('Box plot de edad')
plt.show()
```

```python
#Q2 Box plot de median Peso edad s2
pt = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'edad',
                        aggfunc = 'median',
                        columns = 'sexo',
                    margins= True
                    
                      
                      )
pt
```

```python

sns.boxplot(x = df['peso'],
           ).set_title('Box plot de peso')
plt.show()
```

```python
#Q3 Box plot de median Peso sexo 1
pt = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'cintura',
                        aggfunc = 'median',
                        columns = 'sexo',
                    margins= True
                    
                      
                      )
pt
```

```python
#Q4
pt = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'ropa',
                        aggfunc = 'sum',
                        columns = 'sexo',
                    margins= True
                    
                      
                      )
pt
```
